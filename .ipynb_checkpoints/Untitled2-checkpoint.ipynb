# Python script to convert the Employment_taxes and Turnover columns

## 1. Import pandas, then import the csv 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Estonia Startup Database Q1 2022 export.csv',delimiter=';', skiprows=0, low_memory=False)
df
```


## 2. Creation of a function that strips the column values of the characters 'K' and 'M', then converts the values to the corresponding numerical value

```python
df_employment_taxes_M = df[df.Employment_taxes.str.contains('M')]
df_employment_taxes_K = df[df.Employment_taxes.str.contains('K')]
df_Turnover_M = df[df.Turnover.str.contains('M')]
df_Turnover_K = df[df.Turnover.str.contains('K')]
```

```python
def converter(x):
    if 'M' in x:
        return f"{(float(x.strip('M'))*1000000):,.2f}"
    elif 'K' in x:
        return f"{(float(x.strip('K'))*1000):,.2f}"
```

```python
df.Employment_taxes = df.Employment_taxes.apply(converter)
```

```python
df.Turnover = df.Turnover.apply(converter)
```

## 3. Export the modified csv file

```python
df.to_csv('df_csv.csv')
```
